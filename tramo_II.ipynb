### Dispersión cromática

Se produce debido a que diferentes longitudes de onda, se propagan a diferentes velovidades en la fibra.

#### Dispersión cromática para diferentes longitudes de onda

$$D(\lambda) = \frac{S}{4}(\lambda - \frac{(\lambda_{ZD})^4}{\lambda^3}) \frac{ps}{nm·km}$$

> Conociendo $\lambda_{ZD}$

donde $\lambda_{ZD}$ es la longitud de onda a la que se reporta **dispersión cero** y $S$ es la pendiente de la curva de dispersion en $\lambda_{ZD}$.

* $S = 0.092 \frac{ps}{nm^2·km}$
* Rango de medición de la pendiente: *(se calcula más adelante)*  1300 - 1324 nm

![](https://i.imgur.com/VArt6KC.png)

*Dispersión cromática vs longitud de onda*

```python
for i in range (1,4):
    print('λ' + str(i) + ': ', λs[str(i)]['longitud'])
    # lambda = c/f
    f = λs[str(i)]['longitud'].to('hertz')
    nm = c_luz / f 
    nm = nm.to('nanometer')
    λs[str(i)]['nm'] = nm
    print(nm)
    print()
```

```python
"""
DISPERSIÓN CROMÁTICA PARA DIFERENTES λ
=======================================
"""
def calculo_dispersion_cromatica_km(s, λ_zd, λ_):
    s = s.to_root_units()
    return (s/4) * (λ_ - (λ_zd**4 / λ_**3))
```

```python
print('FIBRA')
print('=====')
print('Pendiente de la dispersión de la fibra: ', fibra['s'])
print('Rango de medición de la pendiente: ', fibra['rango_s'][0], fibra['rango_s'][1])

λs_nm = []
[s, λ_zd] = [fibra['s'], sum(fibra['rango_s']) / len(fibra['rango_s'])]
fibra['dispersiones'] = []
for i in range (1,4):
    λ_nm = λs[str(i)]['nm']
    λs_nm.append(λ_nm)
    dispersion = calculo_dispersion_cromatica_km(s, λ_zd, λ_nm)
    dispersion = dispersion.to('picosecond/(nanometer*kilometer)')
    print('λ' + str(i) + ': ', dispersion)
    fibra['dispersiones'].append(dispersion)
```

Se aproxima la curva de dispersión cromática de la fibra compensadora a una función lineal, asumiento como puntos que pertenecen a la recta, los rangos de medición provistos como dato, contando además con el valor de la pendiente, se determina la función que responde a la ecuación $y = a·x + b$

```python
# Ecuacion de la pendiente de dispersión de fibra de compensación

# y = Sx + b 
# Dc = S(1550) + b
# b = Dc - S(1550)
s = fibra_compensadora['s'].magnitude
dispersion_cromatica = fibra_compensadora['dispersion_cromatica'].magnitude
b = dispersion_cromatica - (s*1550)

print('y = ', s , 'x +', b)
print()
fibra_compensadora['dispersiones'] = []
for i in range (1,4):
    λ = λs[str(i)]['nm']
    dispersion = fibra_compensadora['s'].magnitude * λ.magnitude + b
    dispersion = dispersion * ureg('picosecond/(nanometer*kilometer)')
    print('λ' + str(i) + ': ', dispersion)
    fibra_compensadora['dispersiones'].append(dispersion)
```

$$L_C = \frac{D(\lambda) * L }{- D_C(\lambda)}$$

La cantidad de empalmes necesaria para cada tramo se calcula según la siguiente expresión:

donde $L$ es la longitud del tramo en cuestión, $l$ es la longitud de la fibra.

## OSNR

Se calcula la OSNR por nodo así como acumuladas según:

$$OSNR_{dB} = P_{in} - 10log_{10}(hf1000) + 10log_{10}(B_o) + NF$$

$$OSNR_{veces} = 10^{\frac{OSNR_{dB}}{10}}$$

**OSNR acumulada de N etapas**

$$\frac{1}{OSNR_{veces}} = \frac{1}{OSNR_{veces 1}} + ... + \frac{1}{OSNR_{veces N}}$$

```python
"""
Veces <-> dB
==========
"""

def veces_to_dB(veces):
    if veces == 0:
        return 0
    else:
        return 10 * math.log10(veces)

def dB_to_veces(dB):
    if dB == 0:
        return 0
    else:
        return 10**(dB/10)
```

```python
def calculo_OSNR(Pin, λ):
    c = 3*(10**8)
    h = 6.63 * 10**(-34)
    NF = 5 
    f = (c / λs[str(λ)]['nm'].to_base_units().magnitude)
    Bo = unit_to_dimensionless(λs[str(λ)]['r_bit'].to_base_units())
    OSNR_dB = round(Pin - 10*math.log10(h * f * 1000) - 10* math.log10(Bo) - NF ,2)
    return OSNR_dB
```

```python
df['p_entrada'] = 4 + df['atenuacion_fibra_total_db']
df['p_entrada_2'] = -17
```

```python
df['osnr_ampli_1_λ_1'] = df['p_entrada'].apply(calculo_OSNR, args=[1])
df['osnr_ampli_2_λ_1'] = df['p_entrada_2'].apply(calculo_OSNR, args=[1])

df['osnr_ampli_1_λ_2'] = df['p_entrada'].apply(calculo_OSNR, args=[2])
df['osnr_ampli_2_λ_2'] = df['p_entrada_2'].apply(calculo_OSNR, args=[2])

df['osnr_ampli_1_λ_3'] = df['p_entrada'].apply(calculo_OSNR, args=[3])
df['osnr_ampli_2_λ_3'] = df['p_entrada_2'].apply(calculo_OSNR, args=[3])
```

#### Cálculo de ensanchamiento por PMD

Para cada tramo:

$$ \Delta \tau = PMD * \sqrt L $$

Para N tramos:

$$ \Delta \tau^2  = (PMD * \sqrt L)^2_1 + ... + (PMD * \sqrt L)^2_N $$

```python
"""
dTau por PMD
============
dtau = PMD*sqrt(L)
dtau^2 = sum dtaui^2
"""
def calculo_pdm_tramo(L):
    return( fibra['pdm'] * math.sqrt(L) )

def calculo_pdm_compensadora(L):
    return( fibra_compensadora['pdm'] * math.sqrt(L) )
```

```python
df['ensanchamiento_por_pmd_tramo'] = df['distancia_nodo_anterior'].apply(unit_to_dimensionless).apply(calculo_pdm_tramo)
df['ensanchamiento_por_pmd_compensadora'] = df['distancia_nodo_anterior'].apply(unit_to_dimensionless).apply(calculo_pdm_compensadora)
df['ensanchamiento_por_pmd_tota_cuadrado'] = df['ensanchamiento_por_pmd_tramo']**2 + df['ensanchamiento_por_pmd_compensadora']**2
```

```python
pdm_acumulada = []
acumulador = 0
for i, roadm in enumerate(df.roadm_lambda_1):
    if roadm == 'ASOut' or roadm == 'ASOut, SDOut':
        if 'nan' in str(df.iloc[i]['ensanchamiento_por_pmd_tota_cuadrado']):
            pass
        else:
            acumulador += df.iloc[i]['ensanchamiento_por_pmd_tota_cuadrado']
        while roadm == 'STOut':
            i += 1
            acumulador += df.iloc[i]['ensanchamiento_por_pmd_tota_cuadrado']
        i += 1
        acumulador += df.iloc[i]['ensanchamiento_por_pmd_tota_cuadrado']
        pdm_acumulada.append((math.sqrt(acumulador.magnitude), df.iloc[i]['localidad']))
pdm_acumulada
```

## Análisis de los resultados

Habiendo calculado las longitudes de los módulos compensadores de cada nodo, los valores de **dispersión cromática** para los tramos de fibra y de fibra compensadora, se contempla el caso *más desafortunado*, que corresponde al tramo de mayor extensión, para cada longitud de onda. El *peor* resultado cumple el rango de trabajo impuesto por el receptor.

Los valores de potencia reportados en los puntos más importantes, están contenidos en el rango especificado por el fabricante del amplificador para trabajar por encima de una OSNR deseable, pero por debajo de una figura de ruido significativamente mayor.

Contemplando los valores de OSNR de los elementos amplificadores y transmisores de cada etapa, los acumulados para cada longitud de onda, se compara la peor de las figuras de emérito con el requerimiento impuesto por el receptor, siendo mayor que éste.

Las longitudes efectivas reportadas para los casos más desafortunados superan en 13 y 6 veces la longitud no lineal, para la fibra de tramo y fibra compensadora.

El ensanchamiento debido a la dispersión por modo de polarización no supera el permitido según parámetro de diseño de 10% de tiempo de bit.

## CONCLUSIÓN

A partir de la actividad propuesta para el diseño y análisis del tramo II de la red federal de fibra óptica de Arsat, habiendo elegido los elementos de la red, realizado los cálculos de diseño y análisis de magnitudes, procedimientos y fenómenos en la fibra óptica, así como expuesto visualizaciones que facilitan la toma de desiciones, se establecen las siguientes conclusiones:

La **dispersión cromática** debida a la fibra óptica puede compensarse sin alcanzar límites de ensanchado acumulados provistos por el fabricante del receptormediante módulos compensadores para señales cuya longitud de onda corresponde a 193.9THz, y parcialmente para las longitudes restantes.

La longitud de fibra óptica de los tramos en cuestión no superan las **longitudes críticas** asociadas a los dos tipos de fibra óptica involucrados a partir de las que se debe contemplar la presencia de **SPM** *(Auto Modulación de fase)*.

Para garantizar los niveles de potencia requeridos en los dispositivos presentes en los nodos, es necesario disponer dos amplificadores por nodo, el primero de ellos luego del módulo compensador, mientras que el segundo amplifica la señal inmediatamente antes de propagarse por la fibra hacia el nodo siguiente.

El ensanchamiento producido en la señal por **PMD** *(Dispersión por el modo de polarización)* estocástica no supera el límite impuesto.
